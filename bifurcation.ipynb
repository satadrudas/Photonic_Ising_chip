#Single spin bifurcation based on initial sign

```python
# For a fixed alpha

alpha=1.5
a=15



p1=[]
for j in range(30):
    a=np.power(np.cos(alpha*a),2)
    p1.append(a)

x=np.arange(len(p1))
plt.scatter(x,p1)
plt.plot(x,p1)
plt.title("The value converges to a value(s) independent of the input")
### This happens because for a being either positive or negative would yeild the same value
```

#Multiple spin bifurcation

Here the influencer is only the gaussian noise, but when we include the beta as well, the net result influences the final sign of the spin

##Cosine 

```python
#SWEEPING ACROSS ALPHA


N=100 #number of spins
a= np.zeros([N,1])
traj=a
N_iter=50

alpha_min=0
alpha_max=2
alpha_step=0.1

p1=[]
for alpha in np.arange(alpha_min,alpha_max,alpha_step):
  a= np.zeros([N,1])
  for j in range(N_iter):
    a=np.cos(alpha*a-np.pi/2+np.random.normal(scale=0.05,size=(N,1))) ### works when -alpha
    traj = np.c_[traj,a]

```

```python
x=np.arange(((alpha_max-alpha_min)/alpha_step)*N_iter+1)

plt.rcParams["figure.figsize"] = [20, 4]
plt.rcParams["figure.autolayout"] = True

for i in range(N):
  plt.plot(x,traj[i][:])


## every break in the spplit is basically the start of a new alpha
#this is a concatenated graph
```

# COSINE SQUARED BIFURCATION

```python
a=-1122

b1=[]
x_ticks=[]
for i in np.arange(0.9,2,0.01):
    for j in range(np.random.randint(10,20)):
        a=np.power((np.cos(i*a)),2)
        b1.append(a)
    x_ticks.append(str(np.round(i,decimals=4)))

x=np.arange(len(b1))
#plt.figure(figsize=(20,5))
plt.scatter(x,b1,s=5)
#plt.plot(x,b1)
plt.xlabel('Gain Coefficient',size='large')
plt.ylabel("Convergence value for the given gain")
#plt.xticks(range(len(x_ticks)),x_ticks, size='small',rotation='vertical')
```

```python
#cosine squared bifurcation 
a=-0.9

b1=[]
x_ticks=[]
for i in np.arange(0.9,2,0.01):
    for j in range(12):
        a=np.power((np.cos(i*a+np.pi/4)),2)-0.5
        b1.append(a)
    x_ticks.append(str(np.round(i,decimals=4)))

x=np.arange(len(b1))
#plt.figure(figsize=(20,5))
plt.scatter(x,b1,s=5)
plt.xlabel('Gain Coefficient',size='large')
plt.ylabel("Convergence value for the given gain")
#plt.xticks(range(len(x_ticks)),x_ticks, size='small',rotation='vertical')
```

BASICALLY WHEN a_(i)>0, a_(i+1)<0 and they keep oscillating

#COSINE BIFURCATION

```python
#cosine bifurcation 

a=-0.7
b2=[]
x_ticks=[]

for i in np.arange(0.9,2,0.01):
    for j in range(51):
        a=np.cos(i*a)
        b2.append(a)
    x_ticks.append(str(np.round(i,decimals=4)))

x=np.arange(len(b2))
#plt.figure(figsize=(20,5))
plt.scatter(x,b2,s=5)
plt.xlabel('Gain Coefficient',size='large')
plt.ylabel("Convergence value for the given gain")
```
